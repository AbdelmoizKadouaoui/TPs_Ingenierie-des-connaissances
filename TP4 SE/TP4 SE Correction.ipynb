## Exercice 1 : Modélisation des faits et règles

#### 1. Représenter des faits

```python
faits = ["fievre","toux", "douleurs_thoraciques"]
```

#### 2. la base de règles

```python
base_regles = [ {"conditions": ["fievre", "toux"],"conclusion": "grippe"},
                {"conditions": ["grippe", "douleurs_thoraciques"],"conclusion": "infection_respiratoire"},
                {"conditions": ["infection_respiratoire", "essoufflement"],"conclusion": "hospitalisation_conseillee"}
              ]
```

## Exercice 2 : Chaînage avant 

```python
def chainage_avant(init_Faits, regles):

    nouveau_fait = True

    while nouveau_fait:
        nouveau_fait = False
        for regle in regles:
            if all(cond in init_Faits for cond in regle["conditions"]) and regle["conclusion"] not in init_Faits:
                init_Faits.append(regle["conclusion"])
                nouveau_fait = True

    return init_Faits
```

### 1. Cas-1 que déduit le système ? 

```python
res = chainage_avant(["fievre", "toux"], base_regles)
print("Faits :", '["fievre", "toux"]', "\nLe système déduit : ",res[-1])
```

### 2. Cas-2 quels faits finaux obtient-on ?

```python
res = chainage_avant(["fievre", "toux", "douleurs_thoraciques", "essoufflement"], base_regles)
print("Les Faits :", '["fievre", "toux", "douleurs_thoraciques", "essoufflement"]', "\nLes faits finaux : ",res)
```

### 3. Cas-3 y a-t-il des règles déclenchées ? 

```python
res = chainage_avant(["fievre"], base_regles)
print("Les Faits :", '["fievre"]', "\nLes faits finaux : ",res)
```

Reponse : Non, aucune règle ne se déclenche car : les faits finaux ne changeant pas 

## Exercic 3 : Interaction avec le système 

```python
def peut_on_deduire(FaitInit,Target, base_regles):

    res = chainage_avant(FaitInit, base_regles)
    if Target in res:
        print(f"Oui, {Target} est déduite.")
        return
    print(f"Non, le système ne peut pas conclure {Target}.")
    return
```

#### Test 1 : ["fievre", "toux"] => "grippe"

```python
peut_on_deduire(["fievre", "toux"], "grippe", base_regles)
```

#### Test 2 : "toux", "essoufflement" => "infection_respiratoire"

```python
peut_on_deduire(["fievre","essoufflement"],"infection_respiratoire", base_regles)
```

#### Test 3 : "fievre", "toux", "douleurs_thoraciques" => "infection_respiratoire"

```python
peut_on_deduire(["fievre", "toux", "douleurs_thoraciques"],"infection_respiratoire",base_regles)
```
